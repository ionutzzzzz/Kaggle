```python
def show_corr_heatmap(df, title):
    
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    plt.figure(figsize = (15, 10))
    plt.title(title)
    # sns.heatmap(corr, annot = False, linewidths=.5, fmt=".2f", square=True, mask = mask, cmap=cmap_2)
    if df.shape[1] < 25:
        sns.heatmap(corr, annot=True, linewidths=.5, fmt=".2f", square=True, mask=mask, cmap=cmap_2)
    else:
        sns.heatmap(corr, annot=False, linewidths=.5, square=True, mask=mask, cmap=cmap_2)

    plt.show()
```

```python
def data_description(df):
    print("Data description")
    print(f"Total number of records {df.shape[0]}")
    print(f'number of features {df.shape[1]}\n\n')
    columns = df.columns
    data_type = []
    
    # Get the datatype of features
    for col in df.columns:
        data_type.append(df[col].dtype)
        
    n_uni = df.nunique()
    # Number of NaN values
    n_miss = df.isna().sum()
    
    names = list(zip(columns, data_type, n_uni, n_miss))
    variable_desc = pd.DataFrame(names, columns=["Name","Type","Unique levels","Missing"])
    print(variable_desc)
```

```python
def cross_val(model, 
            label, 
            feat=initial_features, 
            fit_params={}, 
            X = train_df, 
            y = test_df):
    
    start = datetime.datetime.now()
    scs = []
    oof_preds = np.full(len(X), 'x', dtype=str)
    
    for fold, (idx_tr, idx_val) in enumerate(crossval_kf.split(X, targets)):
        model.fit(X.iloc[idx_tr][feat], targets[idx_tr] == 1, **fit_params)
        X_val = X.iloc[idx_val][feat]
        y_val = targets[idx_val] == 1
        y_pred = model.predict(X_val)
        score = matthews_corrcoef(y_val, y_pred)
        del y_val
        
        print(f"{blue}Fold {fold}:{res} {red}Score: {score:.5f}{res}")
        scs.append(score)
        oof_preds[idx_val] = np.where(y_pred, 1, 0)
        break
            
    time_past = datetime.datetime.now() - start
    if not (oof_preds == 'x').any():
        mcrr = matthews_corrcoef(targets, oof_preds)
        logloss = log_loss(targets, oof_preds)   
        print(f"{cyn}Overall: {mcrr:.5f}{res}"
              f"{cyn}logloss: {logloss:.5f} {label}{res}"
              f"{yllw} {int(np.round(time_past.total_seconds() / 60))}:{int(np.round(time_past.total_seconds() % 60))} min{res}")
    else:
        print(f"{cyn}--{label} time: {res}"
              f"{yllw} {int(np.round(time_past.total_seconds() / 60))}:{int(np.round(time_past.total_seconds() % 60))} min{res}")
        
    oof[label] = oof_preds
    fold_scores[label] = scs
    
    if TEST_BOOL:
        X_test = X[feat]
        y_test = targets == 1

        model.fit(X_test, y_test, **fit_params)
        y_pred = model.predict(y[feat])
        test_pred[label] = np.where(y_pred, 1, 0)
```

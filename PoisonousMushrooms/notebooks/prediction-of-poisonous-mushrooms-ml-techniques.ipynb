## 1.5. Literature Review
<ol style="text-align: justify;">
    <li>Pranjal et al. address the problem of distinguishing between edible and poisonous mushrooms using machine learning. They tested various classifiers, including KNN, Logistic Regression, and Decision Trees, based on mushroom texture features. The Support Vector Machine (SVM) classifier outperformed the others, achieving a 76.6% accuracy [1]</li>

<li>Jahan R. et al. highlight the critical issue of mushroom poisoning, which can be fatal if toxic mushrooms are consumed unknowingly. Despite the severe health risks, research on mushroom identification, especially in the context of Bangladeshi mushrooms, is limited. The authors note that while some studies have attempted to address this problem using supervised learning techniques like SVM, Decision Trees, Random Forest, KNN, and Artificial Neural Networks (ANN), ANN has consistently achieved the highest accuracy. Their work aims to compile and review the most recent and accurate studies in this field [2].</li>
    
<li>Kousalya et al. explores the classification of mushrooms as edible or poisonous using data mining techniques. The study compares four popular algorithms: Naive Bayes, Decision Tree (C4.5), Support Vector Machine (SVM), and Logistic Regression, using the WEKA tool for analysis. Testing was conducted on Lepiota and Agaricus mushroom datasets sourced from Kaggle. Among the algorithms, the C4.5 Decision Tree achieved the highest accuracy of 93.34% and also demonstrated the fastest processing speed, outperforming the other methods [3].</li>
    
<li>Tutuncu et al. address the challenge of distinguishing between edible and poisonous mushrooms, a task that typically requires expert knowledge due to the subtle differences between species. To automate this classification, the study utilized the Mushroom dataset from the UC Irvine Machine Learning Repository, analyzing 22 features using four machine learning algorithms: Naive Bayes, Decision Tree, Support Vector Machine, and AdaBoost. The models achieved classification accuracy rates of 90.99%, 98.82%, 99.98%, and 100%, respectively, with AdaBoost proving to be the most accurate, successfully distinguishing between edible and poisonous mushrooms with 100% accuracy [4].</li>
    
<li>Tarawneh et al. emphasize the importance of identifying poisonous mushrooms due to their prevalence and nutritional significance. They note that existing research on mushroom classification primarily focuses on applying individual machine learning techniques, such as Naive Bayes, Decision Tree, and SVM, with varying levels of accuracy. In contrast, their study proposes an integrated model that combines the decisions of the most accurate algorithms into a single decision, rather than treating them individually. Using the Mushroom dataset from the UCI repository, their integrated model achieved a 94% accuracy, outperforming other individual techniques [5].</li>
</ol>

## 2.5. Randomness of Target Class Distribution

```python
import random
import matplotlib.patches as mpatches

# Function to randomly select 3000 entries and return the subset with its index range
def get_random_subset(df, subset_size=3000):
    start_idx = random.randint(0, len(df) - subset_size)
    return df.iloc[start_idx:start_idx + subset_size], (start_idx, start_idx + subset_size)

# Function to plot the class distribution
def plot_class_distribution(ax, data, title, range_info):
    # Iterate through the 'class' column and plot vertical lines
    for i, value in enumerate(data['class']):
        if value == 'e':
            ax.axvline(x=i, color='white', linewidth=0.5)  # White line for class 'e' (edible)
        elif value == 'p':
            ax.axvline(x=i, color='black', linewidth=0.5)  # Black line for class 'p' (poisonous)
    
    # Add legend
    blue_patch = mpatches.Patch(color='white', label='Class e (Edible)')
    magenta_patch = mpatches.Patch(color='black', label='Class p (Poisonous)')
    ax.legend(handles=[blue_patch, magenta_patch])
    
    ax.set_title(f'{title} ({range_info[0]} - {range_info[1]})')
    ax.set_xlabel('Data Entry Index')
    ax.set_ylabel('Class')

# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Generate and plot four random subsets
for i, ax in enumerate(axs.flat):
    subset, range_info = get_random_subset(df_train)
    plot_class_distribution(ax, subset, f'Random Subset {i+1}', range_info)

plt.tight_layout()
plt.show()
```

## 3.3 Handle Outliers

<h3>Remove Outliers in Train Set</h3>

```python
#Get numerical features (excluding 'id' column if present)
numerical_features = df_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'id' in numerical_features:
    numerical_features.remove('id')

#Apply Winsorization (capping and flooring) to handle outliers
for feature in numerical_features:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_train[feature].quantile(0.25)
    Q3 = df_train[feature].quantile(0.75)
    
    # Calculate Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    # Determine lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap values below the lower bound and above the upper bound
    df_train[feature] = np.where(df_train[feature] < lower_bound, lower_bound, df_train[feature])
    df_train[feature] = np.where(df_train[feature] > upper_bound, upper_bound, df_train[feature])

#Verification - Print summary statistics after Winsorization
print("Summary statistics after Winsorization:")
print(df_train[numerical_features].describe())
```

# 5. Machine Learning Techniques

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBClassifier
from sklearn.metrics import matthews_corrcoef
```

## 5.1. Data Preprocessing for Input to Model

```python
# Dropping 'id' column from both training and testing datasets
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
# Identify categorical features, excluding the target column 'class'
categorical_features = df_train.drop('class', axis=1).select_dtypes(include=['object', 'category']).columns

# Initialize one-hot encoder with the updated parameter
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Fit and transform training data
X_train_encoded = encoder.fit_transform(df_train[categorical_features])
X_test_encoded = encoder.transform(df_test[categorical_features])

# Convert the encoded features back to DataFrames with meaningful column names
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out(categorical_features))
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out(categorical_features))

# Combine one-hot encoded categorical features with numerical features
X_train_final = pd.concat([X_train_encoded_df, df_train.drop(categorical_features, axis=1).reset_index(drop=True)], axis=1)
X_test_final = pd.concat([X_test_encoded_df, df_test.drop(categorical_features, axis=1).reset_index(drop=True)], axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Separate features and target in the training set
X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train_final.drop('class', axis=1), 
    X_train_final['class'].apply(lambda x: 1 if x == 'p' else 0),  # Convert target to binary (1 for 'p', 0 for 'e')
    test_size=0.2, 
    random_state=42
)
```

## 5.2. CatBoost Classifier

```python
from catboost import CatBoostClassifier

# Initialize the CatBoost model with GPU acceleration
model = CatBoostClassifier(
    iterations=2300,  # Number of boosting iterations
    learning_rate=0.03,  # Learning rate for finer updates
    depth=12,  # Depth of the trees to capture more complex patterns
    l2_leaf_reg=3,  # L2 regularization parameter to control overfitting
    loss_function='Logloss',  # Loss function
    eval_metric='MCC',  # Supported evaluation metric
    task_type='GPU',  # Use GPU for training
    devices='0',  # Specify the GPU device (e.g., '0' for the first GPU)
    verbose=100  # Print progress every 100 iterations
)

# Train the model
model.fit(
    X_train_split,  # Training features
    y_train_split,  # Training target
    eval_set=[(X_val, y_val)],  # Validation set for early stopping
    early_stopping_rounds=100  # Stop early if no improvement
)
```

## 5.3. Validation

```python
from sklearn.metrics import matthews_corrcoef

# Predict on validation set (use CPU for simplicity)
y_val_pred = model.predict(X_val)

# Calculate Matthews correlation coefficient (MCC) on validation set
mcc = matthews_corrcoef(y_val, y_val_pred)
print(f'MCC on validation set: {mcc:.4f}')
```

## 5.4. Test Predictions Submission

```python
# Predict on the test set (use CPU for simplicity)
y_test_pred = model.predict(X_test_final)

# Convert numerical predictions back to class labels ('p' for poisonous, 'e' for edible)
y_test_pred_class = ['p' if pred == 1 else 'e' for pred in y_test_pred]
```

```python
def checking(df):
  total = len(df)
  null_df = pd.DataFrame(df.isnull().sum(), columns=['#NULLS'])
  null_df['%NULLS'] = round((null_df['#NULLS']/total)*100, 5)
  null_df['#Unique_Valus'] = df.nunique()
  cat_cols = [col for col in df.columns if df[col].dtype == 'object']
  uniques = []
  for col in df.columns:
    if col in cat_cols:
      uniques.append(set(df[col].dropna()))
    else:
     uniques.append(df[col].max() - df[col].min())
  null_df['Unique_Values'] = uniques
  return null_df
```

```python
def outliers_detection(df, q1_coef=0.25, q3_coef=0.75):
  data_out = df.copy()
  outlier_counts = {}

  cat_cols = [col for col in data_out.columns if data_out[col].dtype == 'object']
  num_cols = [col for col in data_out.columns if col not in cat_cols]

  for col in num_cols:
    q1=data_out[col].quantile(q1_coef)
    q3=data_out[col].quantile(q3_coef)
    iqr = q3 - q1
    outlier_counts[col]=[len(data_out[(data_out[col]<q1-1.5*iqr) | (data_out[col]>q3+1.5*iqr)])]

  for col in cat_cols:
    value_counts = data_out[col].value_counts()
    threshold = 0.05*len(data_out)
    rare_values = value_counts[value_counts<threshold].index
    outlier_counts[col]=[len(data_out[data_out[col].isin(rare_values)])]

  outlier_df = pd.DataFrame(outlier_counts).T
  outlier_df.columns = ['Number of Outliers']
  return outlier_df
```

```python
def find_new_categories(generated_dataset, main_dataset, col):
  main_unique_values = set(main_dataset[col].dropna().unique())
  outliers_in_col = generated_dataset[~generated_dataset[col].isin(main_unique_values)]
  return outliers_in_col['class']
```

```python
cat_cols = [col for col in df_train.columns if df_train[col].dtype == 'object']
new_category = pd.DataFrame(columns=['Column', 'e_class', 'p_class'])

for col in cat_cols:
    new_category_rows = find_new_categories(df_train, main_dataset, col)
    e_class_count = new_category_rows[new_category_rows == 'e'].count()
    p_class_count = new_category_rows[new_category_rows == 'p'].count()
    temp_concat = pd.DataFrame({'Column': [col], 'e_class': [e_class_count], 'p_class': [p_class_count]})
    new_category = pd.concat((new_category, temp_concat), axis=0)

new_category
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns

tem_df1 = df_train[['cap-diameter', 'stem-height']]
tem_df2 = df_train[['cap-diameter', 'stem-width']]
tem_df3 = df_train[['stem-height', 'stem-width']]

X1 = tem_df1.fillna(0)
X2 = tem_df2.fillna(0)
X3 = tem_df3.fillna(0)

# K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels1 = kmeans.fit_predict(X1)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels2 = kmeans.fit_predict(X2)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels3 = kmeans.fit_predict(X3)

# Visualize clustering results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='cap-diameter', y='stem-height', hue=kmeans_labels1, data=tem_df1, ax=axes[0])
axes[0].set_title('K-Means Clustering 1')

sns.scatterplot(x='cap-diameter', y='stem-width', hue=kmeans_labels2, data=tem_df2, ax=axes[1])
axes[1].set_title('K-Means Clustering 2')

sns.scatterplot(x='stem-height', y='stem-width', hue=kmeans_labels3, data=tem_df3, ax=axes[2])
axes[2].set_title('K-Means Clustering 3')

plt.show()
```

```python
# Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels1 = gmm.fit_predict(X1)

gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels2 = gmm.fit_predict(X2)

gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels3 = gmm.fit_predict(X3)

# Visualize clustering results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x='cap-diameter', y='stem-height', hue=gmm_labels1, data=tem_df1, ax=axes[0])
axes[0].set_title('GMM Clustering 1')

sns.scatterplot(x='cap-diameter', y='stem-width', hue=gmm_labels2, data=tem_df2, ax=axes[1])
axes[1].set_title('GMM Clustering 2')

sns.scatterplot(x='stem-height', y='stem-width', hue=gmm_labels3, data=tem_df3, ax=axes[2])
axes[2].set_title('GMM Clustering 3')

plt.show()
```

##Data Distribution

```python
def plot_distribution(generated_dataset, main_dataset, column):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.countplot(y=column, data=generated_dataset, order=generated_dataset[column].value_counts().index, palette="viridis")
    plt.title(f'Distribution of {column} in Generated Data')

    plt.subplot(1, 2, 2)
    sns.countplot(y=column, data=main_dataset, order=main_dataset[column].value_counts().index, palette="viridis")
    plt.title(f'Distribution of {column} in Main Data')

    plt.tight_layout()
    plt.show()
```

```python
cat_cols = [col for col in df_train.columns if df_train[col].dtype == 'object']

for col in cat_cols:
  plot_distribution(df_train, main_dataset, col)
  plt.show()
```

# Outlier handling

```python
def outliers_handeling(df, q1_coef=0.25, q3_coef=0.75, iqr_coef=1.5):
    data_out = df.copy()
    cat_cols = [col for col in data_out.columns if data_out[col].dtype == 'object']
    num_cols = [col for col in data_out.columns if col not in cat_cols]

    for col in num_cols:
        q1=data_out[col].quantile(q1_coef)
        q3=data_out[col].quantile(q3_coef)
        iqr = q3 - q1
        data_out.loc[(data_out[col] < q1 - iqr_coef*iqr), col] = q1 - iqr_coef*iqr
        data_out.loc[(data_out[col] > q3 + iqr_coef*iqr), col] = q3 + iqr_coef*iqr

    for col in cat_cols:
        value_counts = data_out[col].value_counts()
        threshold = 100
        rare_values = value_counts[value_counts<threshold].index
        data_out[col] = data_out[col].apply(lambda x: 'Other' if x in rare_values else x)

    return data_out
```

```python
without_outliers_train = outliers_handeling(df_train)
without_outliers_test = outliers_handeling(df_test)
```

```python
for col in cat_cols:
  plot_distribution(without_outliers_train, main_dataset, col)
  plt.show()
```

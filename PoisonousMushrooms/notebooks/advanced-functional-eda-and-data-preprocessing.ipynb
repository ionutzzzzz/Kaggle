
<h2 id="section1-2" style="font-size: 24px; font-weight: bold;"> 1.2) Correcting Anomalous Data in Categorical Variables</h2>

<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li>In categorical variables, misspellings or inconsistencies can lead to an excessive number of classes.</li>
    <li>We observe that variables with data types such as object and those that appear categorical have too many classes.</li>
    <li>It seems that these categorical variables are classified according to the letters of the alphabet.</li>
    <li>Therefore, let's replace any values that do not contain letters of the alphabet with <strong>NaN</strong>.</li>
    <li>Later, we will fill <strong>NaN</strong> using the KNN (K-Nearest Neighbors) algorithm.</li>
</ul>


```python
df.nunique()
```

```python
def replace_non_alpha_with_nan(df):
    # Get columns with dtype 'object'
    object_columns = [col for col in df.columns if df[col].dtype == 'object']
    
    # Define a list of alphabetic characters (lowercase)
    alphabet_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    
    # Function to check if a value is a single alphabetic character
    def filter_alpha(value):
        if isinstance(value, str):
            # Check if the value is a single alphabetic character and in alphabet_list
            return value if len(value) == 1 and value in alphabet_list else np.nan
        return np.nan

    # Apply the function to each object column
    for col in object_columns:
        df[col] = df[col].apply(filter_alpha)
    
    return df
```

```python
df = replace_non_alpha_with_nan(df)
```

<h2 id="section1-3" style="font-size: 24px; font-weight: bold;"> 1.3) Categorization of Variables</h2>


```python
def grab_col_names(dataframe, cat_th=10, car_th=30):
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(df)
```

```python
cat_cols = [col for col in cat_cols if col not in ["class"]]
num_cols = [col for col in num_cols if col not in ["id"]]
```

<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li><strong>Categorical Variables (cat_cols):</strong> Variables with the data type 'object' or numerical variables with unique values below a threshold (default is 10) are included in this category.</li>
    <li><strong>Numerical Variables (num_cols):</strong> Variables that are not of data type 'object' and do not fall into the category of categorical variables with unique values above a threshold are classified as numerical.</li>

</ul>

<h2 id="section2-1" style="font-size: 24px; font-weight: bold;"> 2.1) Outliers</h2>


<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li><strong>`outlier_thresholds`</strong>: Calculates the lower and upper thresholds for outliers in a column.</li>
    <li><strong>`check_outlier`</strong>: Checks if there are any outliers in a column.</li>
    <li><strong>`replace_with_thresholds`</strong>: Replaces outliers with the defined lower and upper thresholds.</li>
</ul>


```python
def outlier_thresholds(dataframe, col_name, q1=0.01, q3=0.99):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
    
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

<h2 id="section2-2" style="font-size: 24px; font-weight: bold;"> 2.2) Missing Values</h2>


<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li>The <strong>`missing_values_table`</strong> function identifies columns with missing values, calculates their counts and percentages, and prints this information, optionally returning the list of columns with missing values if <strong>`na_name`</strong> is set to <strong>`True`</strong>.</li>
</ul>



```python
def missing_values_table(dataframe, na_name=False):
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]

    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)
    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])
    print(missing_df, end="\n")

    if na_name:
        return na_columns


missing_values_table(df)
```

<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li>Upon examining the missing value table, we see that variables such as <strong>`veil-type`</strong>, <strong>`spore-print-color`</strong>, <strong>`stem-root`</strong>, and <strong>`veil-color`</strong> have very high missing value ratios.</li>
    <li>Variables with missing ratios greater than 80% should be dropped directly.</li>
</ul>



```python
columns_to_drop = ['veil-type', 'spore-print-color', 'stem-root', 'veil-color']
df = df.drop(columns=columns_to_drop)
```

<ul style="font-size: 14px; border: 2px solid black; display: inline-block; padding: 10px; list-style-position: inside;">
    <li>After, columns with missing value ratios below 80% are handled.</li>
    <li>Missing values in numerical columns are imputed using K-Nearest Neighbors (KNN) imputation, which estimates missing values based on neighboring observations.</li>
    <li>Numerical data is then transformed to a normal distribution using Quantile Transformer.</li>
    <li>For categorical columns, missing values are filled with the most frequent value.</li>
    <li>Additionally, categorical data is converted to numerical values using Ordinal Encoding.</li>
    <li>These steps make the dataset more meaningful and suitable for analysis.</li>
</ul>


```python
def handle_missing_data(df, num_cols, cat_cols):

    numeric_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=3))
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
    ])


    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, num_cols),
            ('cat', categorical_transformer, cat_cols)
        ]
    )

    df_transformed = pd.DataFrame(preprocessor.fit_transform(df[num_cols + cat_cols]), columns=num_cols + cat_cols)
    
    print("Missing values after imputation:")
    print(df_transformed.isnull().sum())

    df_final = df[['id', 'class']].join(df_transformed)

    return df_final
```

```python
cat_cols, num_cols, cat_but_car = grab_col_names(df)
cat_cols = [col for col in cat_cols if col not in ["class"]]
num_cols = [col for col in num_cols if col not in ["id"]]
```

```python
df = handle_missing_data(df, num_cols, cat_cols)
```

    <li><strong>Label encoding</strong>, on the other hand, will allow us to represent categorical variables with numerical values efficiently without excessively expanding the dataset's size.</li>
</ul>


```python
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

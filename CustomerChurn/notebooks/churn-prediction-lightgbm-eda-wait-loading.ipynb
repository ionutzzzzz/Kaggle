```python
# ── 3.1  Schema Audit ─────────────────────────────────────────────────────────
def display_schema_audit(df, name='DataFrame'):
    display(HTML(
        f"<div style='font-family:Space Mono,monospace;border-left:3.5px solid {P['p2']};"
        f"border-radius:0 8px 8px 0;padding:10px 18px;margin:18px 0 10px;"
        f"background:var(--card);color:var(--heading);font-size:.84em;font-weight:700;'>"
        f"📋 &nbsp;{name} &nbsp;&middot;&nbsp;"
        f"<span style='color:{P['p1']};'>{len(df):,} rows</span> &nbsp;&times;&nbsp;"
        f"<span style='color:{P['p2']};'>{df.shape[1]} cols</span></div>"
    ))
    display(df.head().style.set_table_styles(TBL_STYLES, overwrite=True).set_caption(f'{name} — Head'))

    summary = pd.DataFrame({
        'dtype': df.dtypes, 'non_null': df.notnull().sum(),
        'unique': df.nunique(), 'missing': df.isnull().sum(),
        'missing_%': (df.isnull().mean() * 100).round(2),
    })
    display(
        summary.style
               .background_gradient(subset=['missing_%'], cmap='YlOrBr', vmin=0, vmax=50)
               .background_gradient(subset=['unique'], cmap='Blues')
               .format({'missing_%': '{:.2f}%'})
               .set_table_styles(TBL_STYLES, overwrite=True)
               .set_caption(f'{name} — Column Report')
    )
    num_df = df.select_dtypes(include=np.number)
    if not num_df.empty:
        display(
            num_df.describe().T.style
                  .background_gradient(subset=['mean', 'std'], cmap='Blues')
                  .format(precision=3)
                  .set_table_styles(TBL_STYLES, overwrite=True)
                  .set_caption(f'{name} — Numeric Summary')
        )
    cat_df = df.select_dtypes(exclude=np.number)
    if not cat_df.empty:
        display(
            cat_df.describe().T.style
                  .set_table_styles(TBL_STYLES, overwrite=True)
                  .set_caption(f'{name} — Categorical Summary')
        )

display_schema_audit(df_train, 'Training Set')
display_schema_audit(df_test,  'Test Set')
```

```python
# ── 3.2  Target Distribution — Donut + Annotated Bar ────────────
def plot_target_distribution(df, target):
    """Donut (class share %) + bar (absolute counts). Reveals class imbalance."""
    counts = df[target].value_counts().reset_index()
    counts.columns = [target, 'count']
    counts['pct'] = (counts['count'] / counts['count'].sum() * 100).round(1)
    imbalance = counts['count'].max() / counts['count'].min()

    fig = make_subplots(
        rows=1, cols=2,
        specs=[[{'type': 'domain'}, {'type': 'bar'}]],
        column_widths=[0.38, 0.62],
        subplot_titles=['Class Share (%)', 'Absolute Count'],
    )
    fig.add_trace(go.Pie(
        labels=counts[target], values=counts['count'], hole=0.58,
        marker=dict(colors=[P['p2'], P['p1']], line=dict(color='white', width=3)),
        textinfo='label+percent',
        textfont=dict(size=13, family='Space Mono, monospace'),
        showlegend=False,
    ), row=1, col=1)
    fig.add_trace(go.Bar(
        x=counts[target], y=counts['count'],
        marker=dict(color=[P['p2'], P['p1']], line=dict(color='white', width=1.5)),
        text=counts.apply(lambda r: f"{r['count']:,}<br>({r['pct']}%)", axis=1),
        textposition='outside', showlegend=False,
    ), row=1, col=2)

    existing = list(fig.layout.annotations)
    existing.append(dict(
        text=f'<b>{len(df):,}</b><br>total', x=0.16, y=0.5,
        xref='paper', yref='paper', showarrow=False,
        font=dict(size=13, family='Space Mono, monospace'),
    ))
    fig.update_layout(
        **PL,
        title_text=f'<b>Target Class Distribution</b>  ·  Imbalance ratio ≈ {imbalance:.1f}×',
        height=420, annotations=existing,
    )
    fig.show()

plot_target_distribution(df_train, TARGET)
```

```python
# ── 3.3  Categorical Features — Count (Train/Test) + Churn-Rate Overlay ───────
def plot_categorical_distributions(train_df, test_df, cat_cols, target='Churn'):
    """
    2-column grid. Each panel:
      Bars : count per category, split Train/Test (covariate-shift check)
      Line : mean churn % per category on twin axis (high-risk level signal)
    """
    train_local = train_df.copy()
    train_local[target] = train_local[target].map({'Yes': 1, 'No': 0}).astype(float)

    n_rows = (len(cat_cols) + 1) // 2
    fig, axes = plt.subplots(n_rows, 2, figsize=(18, n_rows * 5.2),
                             facecolor=PL['paper_bgcolor'])
    axes = axes.flatten()

    for i, col in enumerate(cat_cols):
        ax = axes[i]
        ax.set_facecolor(PL['plot_bgcolor'])
        combined = pd.concat([
            train_df[[col]].assign(Source='Train'),
            test_df[[col]].assign(Source='Test'),
        ], ignore_index=True)
        order = combined[col].value_counts().index

        sns.countplot(x=col, hue='Source', data=combined, ax=ax, order=order,
                      palette={'Train': P['p2'], 'Test': P['p1']},
                      alpha=0.85, saturation=0.92)
        ax.set_title(col, fontweight='bold', pad=10)
        ax.set_xlabel('')
        ax.set_ylabel('Count', fontsize=10)
        ax.tick_params(axis='x', rotation=28, labelsize=9)
        leg = ax.get_legend()
        if leg: leg.set_title('Split', prop={'size': 8})

        churn_means = train_local.groupby(col)[target].mean().reindex(order).fillna(0)
        ax2 = ax.twinx()
        ax2.plot(range(len(churn_means)), churn_means.values * 100,
                 color=P['p2'], marker='D', linestyle='--',
                 linewidth=2, markersize=6, markerfacecolor='white', markeredgewidth=2.5)
        ax2.set_ylabel('Churn Rate %', color=P['p2'], fontsize=9)
        ax2.tick_params(axis='y', labelcolor=P['p2'], labelsize=8)
        ax2.set_ylim(0, 100)
        ax2.spines['top'].set_visible(False)
        ax2.spines['right'].set_color(P['p2s'])
        ax2.set_facecolor('none')
        peak = churn_means.values.argmax()
        ax2.annotate(f"{churn_means.values[peak]*100:.0f}%",
                     xy=(peak, churn_means.values[peak] * 100),
                     xytext=(5, 5), textcoords='offset points',
                     fontsize=8, color=P['p2'], fontweight='bold')

    for j in range(i + 1, len(axes)): fig.delaxes(axes[j])
    fig.suptitle('Categorical Features  \u00b7  Train vs Test Counts  \u00b7  \u25c6\u2500\u2500 Churn Rate %',
                 fontsize=13, fontweight='bold', y=1.01)
    plt.tight_layout()
    plt.show()

plot_categorical_distributions(df_train, df_test, CAT_COLUMNS, TARGET)
```

```python
def plot_numerical_distributions(
    train_df: pd.DataFrame,
    test_df: pd.DataFrame,
    numerical_cols: List[str],
    target_col: str,
) -> None:
    """
    For each numerical feature, render:
    - A styled summary stats table
    - Plotly violin: Train vs Test
    - Plotly violin: by target label
    """
    # Normalise target to readable labels for legend
    train_plot = train_df.copy()
    if train_plot[target_col].dtype == 'object':
        train_plot['_label'] = train_plot[target_col].map({'Yes': 'Churned', 'No': 'No Churn'})
    else:
        train_plot['_label'] = train_plot[target_col].map({1: 'Churned', 0: 'No Churn'})

    for col in numerical_cols:
        # ── Styled stats table ────────────────────────────────────────────────
        stats = pd.DataFrame({
            'Train': train_df[col].describe(),
            'Test':  test_df[col].describe(),
        }).loc[['mean', '50%', 'std', 'min', 'max']]
        stats.index = ['Mean', 'Median', 'Std Dev', 'Min', 'Max']

        display(HTML(f"<p style='font-family:Space Mono,monospace; color:#e63946; font-size:0.85em; margin:18px 0 4px;'>📌 {col}</p>"))
        display(
            stats.style
                 .background_gradient(cmap='RdYlGn', axis=1)
                 .format(precision=3)
                 .set_table_styles([{
                     'selector': 'thead th',
                     'props': [('background', '#21262d'), ('color', '#f0f6fc'),
                               ('font-family', 'Space Mono, monospace'),
                               ('padding', '8px 14px'),
                               ('border-bottom', '2px solid #e63946')],
                 }, {
                     'selector': 'tbody td',
                     'props': [('background', '#161b22'), ('color', '#c9d1d9'),
                               ('padding', '7px 14px'),
                               ('border-bottom', '1px solid #21262d'),
                               ('font-family', 'DM Sans, sans-serif')],
                 }], overwrite=True)
        )

        # ── Combined df ───────────────────────────────────────────────────────
        combined = pd.concat([
            train_df[[col]].assign(Split='Train'),
            test_df[[col]].assign(Split='Test'),
        ], ignore_index=True)

        # ── Violin: Train vs Test ─────────────────────────────────────────────
        fig1 = px.violin(
            combined, x='Split', y=col, color='Split', box=True, points='outliers',
            color_discrete_map={'Train': P['p2'], 'Test': P['p1']},
            title=f'<b>{col}</b>  ·  Train vs Test',
        )
        fig1.update_layout(**PL, height=380, showlegend=False)
        fig1.show()

        # ── Violin: by churn ─────────────────────────────────────────────────
        fig2 = px.violin(
            train_plot, x='_label', y=col, color='_label', box=True, points='outliers',
            color_discrete_map={'Churned': P['p2'], 'No Churn': P['p1']},
            title=f'<b>{col}</b>  ·  Distribution by Churn',
            labels={'_label': 'Churn'},
        )
        fig2.update_layout(**PL, height=380, showlegend=False)
        fig2.show()


plot_numerical_distributions(df_train, df_test, NUM_COLUMNS, TARGET)
```

```python
# ── 3.4  Numerical Features — KDE + Box, split by Churn ───────────────────────
def plot_numerical_by_target(df, num_cols, target_col):
    """
    Per numeric feature, one row with two panels:
      KDE (wide)  : density curves per churn class — reveals shape separation
      Box (narrow): box-whisker per churn class — reveals median shift & outliers
    Stats table (calm blue header) printed above each pair.
    """
    df_tmp = df.copy()
    df_tmp['_lbl'] = (df_tmp[target_col]
                      .map({'Yes': 'Churned', 'No': 'No Churn', 1: 'Churned', 0: 'No Churn'})
                      .astype(str))
    pal = {'Churned': P['p2'], 'No Churn': P['p1']}

    for col in num_cols:
        stats = df_tmp.groupby('_lbl')[col].agg(['mean', 'median', 'std', 'min', 'max'])
        display(HTML(
            f"<p style='font-family:Space Mono,monospace;color:{P['p2']};"
            f"font-size:.84em;margin:20px 0 5px;'>\U0001f4cc  {col}</p>"
        ))
        display(
            stats.style
                 .background_gradient(cmap='Blues', axis=0)
                 .format(precision=2)
                 .set_table_styles(TBL_STYLES, overwrite=True)
                 .set_caption(f'{col} — Stats by Churn Class')
        )

        fig, (ax_kde, ax_box) = plt.subplots(
            1, 2, figsize=(16, 4.5),
            facecolor=PL['paper_bgcolor'],
            gridspec_kw={'width_ratios': [3, 1.4]},
        )
        for a in (ax_kde, ax_box): a.set_facecolor(PL['plot_bgcolor'])

        for label, grp in df_tmp.groupby('_lbl'):
            vals = grp[col].dropna()
            kde  = gaussian_kde(vals)
            xs   = np.linspace(vals.min(), vals.max(), 300)
            ys   = kde(xs)
            ax_kde.plot(xs, ys, color=pal[label], linewidth=2.2, label=label)
            ax_kde.fill_between(xs, ys, alpha=0.14, color=pal[label])
            ax_kde.axvline(vals.median(), color=pal[label],
                           linewidth=1.3, linestyle=':', alpha=0.85)

        ax_kde.set_title(f'{col}  \u00b7  KDE by Churn', fontweight='bold')
        ax_kde.set_xlabel(col); ax_kde.set_ylabel('Density')
        ax_kde.legend(title='Churn', fontsize=9, title_fontsize=9)
        ax_kde.text(0.97, 0.97, '\u22ee dotted = median',
                    transform=ax_kde.transAxes, ha='right', va='top',
                    fontsize=8, color='gray', style='italic')

        sns.boxplot(data=df_tmp, y=col, x='_lbl', ax=ax_box,
                    order=['No Churn', 'Churned'], palette=pal,
                    width=0.45, linewidth=1.4, fliersize=3,
                    medianprops=dict(color='white', linewidth=2.2))
        ax_box.set_title(f'{col}  \u00b7  Box by Churn', fontweight='bold')
        ax_box.set_xlabel(''); ax_box.set_ylabel('')
        ax_box.tick_params(axis='x', labelsize=10)
        plt.tight_layout(); plt.show()

plot_numerical_by_target(df_train, NUM_COLUMNS, TARGET)
```

```python
# ── 3.6  Churn Rate by Key Segment ───────────────────────────
def plot_churn_rate_by_segment(df, segment_cols, target='Churn'):
    """
    Interactive grouped bars: churn % per category value.
    Bar colour gradient: high-churn -> p2, low-churn -> p1.
    """
    df_tmp = df.copy()
    df_tmp[target] = (df_tmp[target].map({'Yes': 1, 'No': 0})
                      if df_tmp[target].dtype == 'object'
                      else df_tmp[target].astype(float))

    c1r, c1g, c1b = _hex2rgb(P['p1'])
    c2r, c2g, c2b = _hex2rgb(P['p2'])

    fig = make_subplots(rows=1, cols=len(segment_cols),
                        subplot_titles=[f'<b>{c}</b>' for c in segment_cols],
                        shared_yaxes=True)

    for j, col in enumerate(segment_cols, 1):
        rates = (df_tmp.groupby(col)[target].mean() * 100
                 ).reset_index().sort_values(target, ascending=False)
        rates.columns = [col, 'churn_rate']
        norm = rates['churn_rate'] / rates['churn_rate'].max()
        colors = [
            f'rgba({int(c2r*t+c1r*(1-t))},{int(c2g*t+c1g*(1-t))},{int(c2b*t+c1b*(1-t))},0.85)'
            for t in norm
        ]
        fig.add_trace(go.Bar(
            x=rates[col], y=rates['churn_rate'],
            marker=dict(color=colors, line=dict(color='white', width=1)),
            text=rates['churn_rate'].round(1).astype(str) + '%',
            textposition='outside',
            hovertemplate=f'<b>%{{x}}</b><br>Churn: %{{y:.1f}}%<extra>{col}</extra>',
            showlegend=False,
        ), row=1, col=j)

    fig.update_layout(
        **PL,
        title_text='<b>Churn Rate (%) by Key Segments</b>  ·  Dark bar = higher risk',
        height=440,
        yaxis=dict(title='Churn Rate (%)', ticksuffix='%'),
    )
    fig.show()

plot_churn_rate_by_segment(df_train, ['Contract', 'InternetService', 'PaymentMethod'], TARGET)
```

```python
# ── 3.7  Churn Funnel ─ Stacked Bar per Segment ──────────────────────────
def plot_churn_funnel(df, segment_cols, target='Churn'):
    """
    Each bar = total customers in a category, split into Churned (p2) vs Retained (p1).
    Shows VOLUME + churn rate simultaneously ─ one view beats two separate charts.
    """
    df_tmp = df.copy()
    df_tmp[target] = df_tmp[target].map({'Yes': 1, 'No': 0}).astype(float)

    fig = make_subplots(rows=1, cols=len(segment_cols),
                        subplot_titles=[f'<b>{c}</b>' for c in segment_cols])

    for j, col in enumerate(segment_cols, 1):
        grp = df_tmp.groupby(col)[target].agg(['sum', 'count']).reset_index()
        grp.columns = [col, 'churned', 'total']
        grp['retained'] = grp['total'] - grp['churned']
        grp = grp.sort_values('total', ascending=False)

        fig.add_trace(go.Bar(
            name='Churned', x=grp[col], y=grp['churned'],
            marker_color=P['p2'], opacity=0.87,
            hovertemplate='<b>%{x}</b><br>Churned: %{y:,}<extra>Churned</extra>',
            showlegend=(j == 1),
        ), row=1, col=j)
        fig.add_trace(go.Bar(
            name='Retained', x=grp[col], y=grp['retained'],
            marker_color=P['p1'], opacity=0.82,
            hovertemplate='<b>%{x}</b><br>Retained: %{y:,}<extra>Retained</extra>',
            showlegend=(j == 1),
        ), row=1, col=j)

    # Create a new dictionary without the 'legend' key from PL to avoid duplication
    pl_without_legend = {k: v for k, v in PL.items() if k != 'legend'}

    fig.update_layout(
        **pl_without_legend,
        barmode='stack',
        title_text='<b>Churn Funnel</b>  ·  Volume + Churn in one view (stack = total customers)',
        height=440,
        legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    )
    fig.show()

plot_churn_funnel(df_train, ['Contract', 'InternetService', 'PaymentMethod'], TARGET)
```

```python
def frequency_encode(df, columns):
    """Each value -> relative frequency in the dataset."""
    out = pd.DataFrame(index=df.index)
    for col in columns:
        if col in df.columns:
            fm = df[col].value_counts(normalize=True).to_dict()
            out[f'{col}_freq'] = df[col].map(fm).fillna(0)
    return out


def engineer_features(X, y, X_test, num_cols, cat_cols, folds, seed):
    """
    10-step feature engineering pipeline.
    Returns: X_train_final, X_test_final, cat_feature_names
    """
    Xc = X.copy(); Xtc = X_test.copy()
    Xf  = pd.DataFrame(index=X.index)
    Xft = pd.DataFrame(index=X_test.index)
    cat_names = []
    all_cols  = num_cols + cat_cols

    STEPS = [
        'Frequency Encoding', 'Target Encoding (OOF)', 'RobustScaler',
        'KBins Discretizer', 'Ordinal Encoding', 'Pairwise Combinations',
        'Interaction Features', 'Polynomial Features', 'Ratio Features', 'Group Aggregates',
    ]
    sep = '  ' + chr(8212) * 53
    print(sep)
    print(f'  Feature Engineering Pipeline  \u00b7  {len(all_cols)} input cols')
    print(sep)

    for idx, step in enumerate(STEPS, 1):
        print(f'  [{idx:02d}/{len(STEPS)}]  {step}')

        if idx == 1:
            Xf  = pd.concat([Xf,  frequency_encode(Xc,  all_cols)], axis=1)
            Xft = pd.concat([Xft, frequency_encode(Xtc, all_cols)], axis=1)

        elif idx == 2:
            yn  = y.map({'Yes':1,'No':0}).astype(float) if y.dtype=='object' else y.astype(float)
            gm  = yn.mean()
            Xt_tr = pd.DataFrame(index=X.index)
            Xt_te = pd.DataFrame(index=X_test.index)
            skf   = StratifiedKFold(n_splits=folds, shuffle=True, random_state=seed)
            for _, (tri, vli) in enumerate(skf.split(X, yn)):
                fmaps = {col: yn.iloc[tri].groupby(X.iloc[tri][col]).mean().to_dict()
                         for col in all_cols if col in X.columns}
                for col in all_cols:
                    if col in X.columns:
                        Xt_tr.loc[vli, f'{col}_target'] = X.iloc[vli][col].map(fmaps.get(col,{})).fillna(gm)
            fm2 = {col: yn.groupby(X[col]).mean().to_dict() for col in all_cols if col in X.columns}
            for col in all_cols:
                if col in X_test.columns:
                    Xt_te[f'{col}_target'] = X_test[col].map(fm2.get(col,{})).fillna(gm)
            Xf  = pd.concat([Xf,  Xt_tr], axis=1)
            Xft = pd.concat([Xft, Xt_te], axis=1)

        elif idx == 3:
            sc = RobustScaler(); sc_cols = [f'{c}_scaled' for c in num_cols]
            Xf  = pd.concat([Xf,  pd.DataFrame(sc.fit_transform(Xc[num_cols]),  columns=sc_cols, index=X.index)],      axis=1)
            Xft = pd.concat([Xft, pd.DataFrame(sc.transform(Xtc[num_cols]),     columns=sc_cols, index=X_test.index)], axis=1)

        elif idx == 4:
            bn = KBinsDiscretizer(n_bins=10, strategy='uniform', encode='ordinal', random_state=seed)
            bc = [f'{c}_bin' for c in num_cols]
            Xf  = pd.concat([Xf,  pd.DataFrame(bn.fit_transform(Xc[num_cols]),  columns=bc, index=X.index).astype(int)],      axis=1)
            Xft = pd.concat([Xft, pd.DataFrame(bn.transform(Xtc[num_cols]),     columns=bc, index=X_test.index).astype(int)], axis=1)
            cat_names.extend(bc)

        elif idx == 5:
            oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
            oc = [f'{c}_ordinal' for c in cat_cols]
            Xf  = pd.concat([Xf,  pd.DataFrame(oe.fit_transform(Xc[cat_cols]),  columns=oc, index=X.index).astype(int)],      axis=1)
            Xft = pd.concat([Xft, pd.DataFrame(oe.transform(Xtc[cat_cols]),     columns=oc, index=X_test.index).astype(int)], axis=1)
            cat_names.extend(oc)

        elif idx == 6:
            for c1, c2 in combinations(cat_cols, 2):
                nm = f'{c1}_x_{c2}'
                Xc[nm]  = Xc[c1].astype(str)  + '_' + Xc[c2].astype(str)
                Xtc[nm] = Xtc[c1].astype(str) + '_' + Xtc[c2].astype(str)
                cat_names.append(nm)
            cc = [c for c in Xc.columns if '_x_' in c]
            Xf  = pd.concat([Xf,  Xc[cc]],  axis=1)
            Xft = pd.concat([Xft, Xtc[cc]], axis=1)

        elif idx == 7:
            Xc['tenure_x_monthly']  = Xc['tenure']  * Xc['MonthlyCharges']
            Xtc['tenure_x_monthly'] = Xtc['tenure'] * Xtc['MonthlyCharges']
            Xc['senior_x_monthly']  = Xc['SeniorCitizen']  * Xc['MonthlyCharges']
            Xtc['senior_x_monthly'] = Xtc['SeniorCitizen'] * Xtc['MonthlyCharges']
            ix = ['tenure_x_monthly', 'senior_x_monthly']
            Xf  = pd.concat([Xf,  Xc[ix]],  axis=1)
            Xft = pd.concat([Xft, Xtc[ix]], axis=1)

        elif idx == 8:
            pi = ['tenure', 'MonthlyCharges', 'TotalCharges']
            po = PolynomialFeatures(degree=2, include_bias=False)
            # Fit the transformer before getting feature names
            po.fit(Xc[pi])
            pn = po.get_feature_names_out(pi)
            Xf  = pd.concat([Xf,  pd.DataFrame(po.transform(Xc[pi]),  columns=pn, index=X.index)],      axis=1)
            Xft = pd.concat([Xft, pd.DataFrame(po.transform(Xtc[pi]), columns=pn, index=X_test.index)], axis=1)

        elif idx == 9:
            eps = 1e-6
            Xc['total_per_tenure']  = Xc['TotalCharges']  / (Xc['tenure'].replace(0, np.nan).fillna(eps)  + eps)
            Xtc['total_per_tenure'] = Xtc['TotalCharges'] / (Xtc['tenure'].replace(0, np.nan).fillna(eps) + eps)
            Xf  = pd.concat([Xf,  Xc[['total_per_tenure']]],  axis=1)
            Xft = pd.concat([Xft, Xtc[['total_per_tenure']]], axis=1)

        elif idx == 10:
            gcols = ['Contract', 'InternetService', 'PaymentMethod']
            new_agg = []
            for g in gcols:
                if g not in Xc.columns: continue
                for nc in num_cols:
                    if nc not in Xc.columns: continue
                    for m in ['mean', 'median', 'std']:
                        cn = f'{nc}_by_{g}_{m}'
                        new_agg.append(cn)
                        Xc[cn]  = Xc.groupby(g)[nc].transform(m).fillna(0)
                        mp = Xc.groupby(g)[nc].agg(m).to_dict()
                        Xtc[cn] = Xtc[g].map(mp).fillna(Xc[cn].mean())
            tn = [c for c in new_agg if c not in X.columns]
            Xf  = pd.concat([Xf,  Xc[tn]],  axis=1)
            Xft = pd.concat([Xft, Xtc[tn]], axis=1)

    for col in cat_names:
        for frame in (Xf, Xft):
            if col in frame.columns:
                frame[col] = frame[col].astype(str).astype('category')

    print(sep)
    print(f'  ✅  {X.shape[1]} raw  →  {Xf.shape[1]} engineered features')
    return Xf, Xft, cat_names


X_train_eng, X_test_eng, cat_feature_names = engineer_features(
    X_raw, y_raw, X_test_raw, NUM_COLUMNS, CAT_COLUMNS, FOLDS, SEED
)

display(HTML(
    '<div style="display:flex;gap:13px;flex-wrap:wrap;margin-top:12px;">'
    + _stat_card('Original Features',   str(X_raw.shape[1]),       P['p2'], '')
    + _stat_card('Engineered Features', str(X_train_eng.shape[1]), P['p1'], '')
    + _stat_card('Training Samples',    f"{X_train_eng.shape[0]:,}", P['neutral'], '')
    + '</div>'
))
```

```python
def train_lgbm_ensemble(X_train, y, X_test, folds, seeds, params):
    """
    Stratified K-Fold x multi-seed LightGBM ensemble.
    Returns: models, test_preds (mean across seeds), oof_preds (mean across seeds)
    """
    early_stop = params.pop('early_stopping_round', None)
    verbose    = params.pop('verbose', -1)
    all_models, oof_per_seed, test_per_seed = [], [], []

    for seed in seeds:
        display(HTML(
            f"<div style='font-family:Space Mono,monospace;background:var(--card);"
            f"border-left:3.5px solid {P['p2']};border-radius:0 8px 8px 0;"
            f"padding:9px 16px;margin:13px 0 5px;color:var(--heading);font-size:.82em;'>"
            f"\U0001f331 &nbsp;Seed <span style='color:{P['p2']};font-weight:700;'>{seed}</span></div>"
        ))
        skf   = StratifiedKFold(n_splits=folds, shuffle=True, random_state=seed)
        oof   = np.zeros(len(X_train))
        t_prd = np.zeros(len(X_test))
        fold_aucs = []

        for fn, (tri, vli) in enumerate(skf.split(X_train, y), 1):
            X_tr, X_val = X_train.iloc[tri], X_train.iloc[vli]
            y_tr, y_val = y.iloc[tri],        y.iloc[vli]
            model = lgb.LGBMClassifier(**{**params, 'random_state': seed, 'verbose': verbose})
            lgb_cats = [c for c in X_tr.columns
                        if X_tr[c].dtype == 'object' or pd.api.types.is_categorical_dtype(X_tr[c])]
            fkw = {'eval_set': [(X_val, y_val)]}
            if early_stop:
                fkw['callbacks'] = [lgb.early_stopping(stopping_rounds=early_stop, verbose=False)]
            if lgb_cats:
                fkw['categorical_feature'] = lgb_cats

            model.fit(X_tr, y_tr, **fkw)
            all_models.append(model)

            vp = model.predict_proba(X_val)[:, 1]
            tp = model.predict_proba(X_test)[:, 1]
            oof[vli] += vp
            t_prd    += tp / folds

            fa  = roc_auc_score(y_val, vp)
            fold_aucs.append(fa)
            bar = chr(9608) * int(fa * 28) + chr(9617) * (28 - int(fa * 28))
            print(f'    Fold {fn}  AUC={fa:.5f}  |{bar}|')

        seed_auc = roc_auc_score(y, oof)
        display(HTML(
            f"<div style='font-family:Space Mono,monospace;display:flex;gap:22px;"
            f"background:var(--bg);border:1px solid var(--border);"
            f"border-radius:8px;padding:10px 18px;margin:5px 0;font-size:.82em;'>"
            f"<span>OOF AUC &nbsp;<b style='color:{P['p2']};'>{seed_auc:.5f}</b></span>"
            f"<span>Mean Fold &nbsp;<b style='color:{P['p1']};'>{np.mean(fold_aucs):.5f}</b></span>"
            f"<span>Std &nbsp;<b style='color:var(--muted);'>±{np.std(fold_aucs):.5f}</b></span>"
            f"</div>"
        ))
        oof_per_seed.append(oof)
        test_per_seed.append(t_prd)

    return all_models, np.mean(test_per_seed, axis=0), np.mean(oof_per_seed, axis=0)


lgbm_params = LIGHTGBM_PARAMS.copy()
trained_models, test_predictions, oof_predictions = train_lgbm_ensemble(
    X_train_eng, y_raw, X_test_eng, folds=FOLDS, seeds=SEEDS, params=lgbm_params
)
```

```python
def find_optimal_threshold(y_true, y_proba, thresholds=np.arange(0.05, 0.96, 0.01)):
    """Grid search for the threshold that maximises Matthews Correlation Coefficient."""
    best_mcc, best_thresh = -np.inf, 0.5
    for t in tqdm(thresholds, desc='Threshold search'):
        mcc = matthews_corrcoef(y_true, (y_proba >= t).astype(int))
        if mcc > best_mcc:
            best_mcc, best_thresh = mcc, t
    display(HTML(
        '<div style="display:flex;gap:13px;flex-wrap:wrap;margin-top:10px;">'
        + _stat_card('Optimal Threshold', f'{best_thresh:.2f}', P['p2'], 'maximises MCC')
        + _stat_card('Best MCC',          f'{best_mcc:.5f}',   P['p1'], 'on OOF predictions')
        + '</div>'
    ))
    return best_thresh, best_mcc


optimal_threshold, best_mcc = find_optimal_threshold(y_raw, oof_predictions)
y_pred_train = (oof_predictions >= optimal_threshold).astype(int)
```

```python
# ── Metrics Scorecard ───────────────────────────
def display_metrics_scorecard(y_true, y_proba, y_pred, model_name='LightGBM'):
    """
    Four metric cards with inline mini progress bar.
    Card accent colour: steel-blue (calm) — green/red only for pass/fail signal.
    """
    scores = {
        'ROC-AUC':  roc_auc_score(y_true, y_proba),
        'Accuracy': accuracy_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'MCC':      matthews_corrcoef(y_true, y_pred),
    }
    icons = {'ROC-AUC': '🎯', 'Accuracy': '✅', 'F1-Score': '⚖️', 'MCC': '📐'}
    # Calm tier: great=steel-blue, ok=slate-gray, needs-work=muted-red
    def _c(v): return '#3d5a80' if v >= 0.80 else ('#6b7d8e' if v >= 0.70 else '#a94442')

    cards = ''.join([
        f"<div style='flex:1;min-width:115px;background:var(--bg);"
        f"border:1px solid var(--border);border-radius:10px;padding:16px 13px;"
        f"text-align:center;border-top:3px solid {_c(v)};'>"
        f"<p style='color:var(--muted);font-size:.7em;margin:0 0 5px;"
        f"text-transform:uppercase;letter-spacing:1px;font-family:Space Mono,monospace;'>{icons[k]} {k}</p>"
        f"<p style='color:{_c(v)};font-family:Space Mono,monospace;"
        f"font-size:1.38em;font-weight:700;margin:0 0 7px;'>{v:.4f}</p>"
        f"<div style='height:5px;border-radius:3px;background:var(--border);'>"
        f"<div style='height:5px;border-radius:3px;width:{min(v*100,100):.0f}%;"
        f"background:{_c(v)};transition:width 0.4s;'></div></div></div>"
        for k, v in scores.items()
    ])

    display(HTML(
        f"<div style='font-family:DM Sans,sans-serif;background:var(--card);"
        f"border:1px solid var(--border);border-radius:12px;padding:20px 22px;margin:12px 0;'>"
        f"<p style='color:var(--heading);font-family:Space Mono,monospace;font-size:.77em;"
        f"text-transform:uppercase;letter-spacing:1.5px;margin:0 0 14px;"
        f"padding-bottom:10px;border-bottom:1px solid var(--border);'>"
        f"📊 &nbsp;{model_name} · OOF Performance</p>"
        f"<div style='display:flex;gap:11px;flex-wrap:wrap;'>{cards}</div>"
        f"<p style='color:var(--sub);font-size:.75em;margin:14px 0 0;'>"
        f"🟦 steel-blue = good (≤80%)  ▪️ slate = ok (70-80%)  🟥 muted-red = needs work (<70%)</p>"
        f"</div>"
    ))
    return pd.DataFrame([scores])


# ── Confusion Matrix ───────────────────────────────────────────────
def plot_confusion_matrix_styled(y_true, y_pred, title='Confusion Matrix'):
    """Heatmap: p1 diagonal (correct), muted-red off-diagonal (errors). Labels: TN/FP/FN/TP."""
    cm = confusion_matrix(y_true, y_pred)
    labels = ['No Churn', 'Churned']
    mid = '#f5f5f5' if PLOT_THEME == 'light' else '#21262d'
    cell_labels = [['TN', 'FP'], ['FN', 'TP']]
    text = [[f'<b>{cm[i,j]:,}</b><br><span style="font-size:11px;opacity:0.7">{cell_labels[i][j]}</span>'
             for j in range(2)] for i in range(2)]
    fig = go.Figure(go.Heatmap(
        z=cm, x=labels, y=labels,
        colorscale=[[0, mid], [0.4, P['p2s']], [1, P['p1']]],
        zmin=0, zmax=cm.max(),
        text=text, texttemplate='%{text}', textfont=dict(size=16),
        showscale=False,
    ))
    fig.update_layout(
        **PL,
        title_text=f'<b>{title}</b>  ·  {P["p1"]}=correct, muted-red=error  ·  TN/FP/FN/TP labelled',
        height=440,
        xaxis=dict(title='Predicted', tickfont=dict(size=13)),
        yaxis=dict(title='Actual',    tickfont=dict(size=13)),
    )
    fig.show()

metrics_df = display_metrics_scorecard(y_raw, oof_predictions, y_pred_train)
plot_confusion_matrix_styled(y_raw, y_pred_train, 'Confusion Matrix — LightGBM OOF')

```

```python
# ── ROC Curve ─────────────────────────────────────────
def plot_roc_curve_styled(y_true, y_proba, title='ROC Curve'):
    """p1 fill + p2 random-chance baseline."""
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    auc = roc_auc_score(y_true, y_proba)
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=fpr, y=tpr, mode='lines',
        line=dict(color=P['p1'], width=2.8),
        fill='tozeroy', fillcolor=P['p1f'],
        name=f'LightGBM  AUC={auc:.4f}',
    ))
    fig.add_trace(go.Scatter(
        x=[0, 1], y=[0, 1], mode='lines',
        line=dict(color=P['p2'], width=1.5, dash='dot'),
        name='Random (AUC=0.50)',
    ))
    fig.add_annotation(
        x=0.72, y=0.22,
        text=f'<b>AUC = {auc:.4f}</b>',
        showarrow=False,
        font=dict(size=14, color=P['p1'], family='Space Mono, monospace'),
        bgcolor=PL['paper_bgcolor'],
        bordercolor=P['p1'], borderwidth=1.5, borderpad=10,
    )

    # Create a new dictionary without the 'legend' key from PL to avoid duplication
    pl_without_legend = {k: v for k, v in PL.items() if k != 'legend'}

    fig.update_layout(
        **pl_without_legend, title_text=f'<b>{title}</b>', height=480,
        xaxis=dict(title='False Positive Rate', range=[0, 1]),
        yaxis=dict(title='True Positive Rate',  range=[0, 1.02]),
        legend=dict(x=0.05, y=0.95),
    )
    fig.show()


plot_roc_curve_styled(y_raw, oof_predictions,     'ROC Curve — LightGBM OOF')

```

```python
# ── Feature Importances ───────────────────────────────────
def plot_feature_importances(models, feature_names, top_n=25):
    """Horizontal bars with smooth p2->p1 gradient by rank."""
    imps = [m.feature_importances_ for m in models if hasattr(m, 'feature_importances_')]
    if not imps: print('No importances available.'); return
    df_imp = (pd.DataFrame({'Feature': feature_names, 'Importance': np.mean(imps, axis=0)})
              .sort_values('Importance', ascending=True).tail(top_n))
    norm = (df_imp['Importance'] - df_imp['Importance'].min())
    norm /= (norm.max() + 1e-9)
    c1r, c1g, c1b = _hex2rgb(P['p1']); c2r, c2g, c2b = _hex2rgb(P['p2'])
    colors = [f'rgb({int(c2r+t*(c1r-c2r))},{int(c2g+t*(c1g-c2g))},{int(c2b+t*(c1b-c2b))})'
              for t in norm.values]
    fig = go.Figure(go.Bar(
        x=df_imp['Importance'], y=df_imp['Feature'], orientation='h',
        marker=dict(color=colors, line=dict(color='rgba(255,255,255,0.3)', width=0.5)),
        hovertemplate='<b>%{y}</b><br>Importance: %{x:.1f}<extra></extra>',
    ))
    fig.update_layout(
        **PL,
        title_text=f'<b>Top {top_n} Feature Importances</b>  ·  {P["p2"]} low → {P["p1"]} high',
        height=max(520, top_n * 24),
        xaxis=dict(title='Mean Gain Importance'),
        yaxis=dict(tickfont=dict(size=10)),
    )
    fig.show()


# ── Execute all evaluation plots ───────────────────────

plot_feature_importances(trained_models, X_train_eng.columns.tolist(), top_n=25)
```

```python
def generate_submission(test_preds, ids, target_column=TARGET, output_filename='submission.csv'):
    """
    Save submission CSV and display:
      - Styled preview table (calm blue-gray header)
      - Probability distribution histogram with threshold lines
      - Stat cards: file, rows, range, mean
    """
    submission = pd.DataFrame({'id': ids, target_column: test_preds})
    submission.to_csv(output_filename, index=False)

    # ── Console logs ────────────────────────────────
    sep = '  ' + chr(8212) * 50
    print(sep)
    print(f'  Submission Report')
    print(sep)
    print(f'  ✅  Saved           : {output_filename}')
    print(f'  📊  Rows             : {len(submission):,}')
    print(f'  📊  Pred range       : [{test_preds.min():.4f}, {test_preds.max():.4f}]')
    print(f'  📊  Mean pred        : {test_preds.mean():.4f}  |  Std: {test_preds.std():.4f}')
    print(f'  📊  % pred churn     : {(test_preds >= 0.5).mean()*100:.1f}%  (threshold = 0.5)')
    print(f'  📊  % pred churn     : {(test_preds >= optimal_threshold).mean()*100:.1f}%  (threshold = {optimal_threshold:.2f}, optimal MCC)')
    print(sep)

    # ── Styled preview table ────────────────────────────────
    display(
        submission.head(12).style
                  .background_gradient(subset=[target_column], cmap='RdYlGn', vmin=0, vmax=1)
                  .format({target_column: '{:.4f}'})
                  .set_table_styles(TBL_STYLES, overwrite=True)
                  .set_caption('Submission Preview — First 12 Rows (RdYlGn = churn probability)')
    )

    # ── Prediction distribution histogram ───────────────────
    fig = go.Figure()
    fig.add_trace(go.Histogram(
        x=test_preds, nbinsx=50,
        marker=dict(
            color=test_preds,
            colorscale=[[0, P['p1']], [0.5, '#e8e8e8'], [1, P['p2']]],
            line=dict(color='white', width=0.4),
        ),
        hovertemplate='Prob: %{x:.3f}<br>Count: %{y}<extra></extra>',
        showlegend=False,
    ))
    fig.add_vline(x=0.5, line=dict(color=P['p2'], width=1.8, dash='dash'),
                  annotation_text='  threshold=0.50', annotation_font_size=11)
    fig.add_vline(x=optimal_threshold,
                  line=dict(color=P['p1'], width=1.8, dash='dot'),
                  annotation_text=f'  optimal={optimal_threshold:.2f} (MCC)',
                  annotation_font_size=11)
    fig.update_layout(
        **PL,
        title_text='<b>Predicted Churn Probability Distribution</b>  ·  Test Set',
        height=380,
        xaxis=dict(title='Predicted Probability', range=[0, 1]),
        yaxis=dict(title='Count'),
    )
    fig.show()

    # ── Stat cards ──────────────────────────────
    display(HTML(
        '<div style="display:flex;gap:13px;flex-wrap:wrap;margin-top:10px;">'
        + _stat_card('File',        output_filename,                      P['p1'],    'saved ✓')
        + _stat_card('Rows',        f"{len(submission):,}",               P['neutral'], 'samples')
        + _stat_card('Pred Range',  f'[{test_preds.min():.3f}, {test_preds.max():.3f}]', P['p2'], 'interval')
        + _stat_card('Mean Pred',   f'{test_preds.mean():.4f}',           '#f59e0b', 'avg probability')
        + '</div>'
    ))
    return submission


submission_df = generate_submission(
    test_preds=test_predictions,
    ids=test_ids,
    target_column=TARGET,
    output_filename='submission.csv',
)
```

```python
def check_data_quality(df, title):
    display(HTML(f"""
        <div style="
          padding: 14px 26px;
          background: linear-gradient(90deg, #8CAEC2, #BB8CC2);
          border-radius: 12px;
          border-left: 7px solid lightblue;
          color: white;
          font-family: 'Inter', sans-serif;
          margin: 42px 0 18px;
          box-shadow: 0 4px 16px rgba(0,0,0,0.08);
          display: flex;
          align-items: center;
          gap: 14px;
        ">
          <h3 style="font-size: 18px; font-weight: 700; margin: 0; color: white;">📋 {title}</h3>
        </div>
    """))

    missing = df.isnull().sum()
    pct_missing = (missing / len(df)) * 100
    health_df = pd.DataFrame({
        'Missing Values': missing,
        'Percentage (%)': pct_missing.round(2)
    })
    report = health_df[health_df['Missing Values'] > 0].sort_values(
        by='Percentage (%)', ascending=False
    )

    if report.empty:
        display(HTML("""
            <div style="
              padding: 12px 18px;
              background: #e8f4fb;
              border-left: 5px solid #8CAEC2;
              border-radius: 8px;
              margin: 8px 0;
              font-size: 13.5px;
              color: #2c5f7a;
              font-weight: 600;
              line-height: 1.6;
              font-family: 'Inter', sans-serif;
            ">
                ✅ <strong>MISSING VALUES</strong><br>
                <span style="font-weight: 400;">No missing values detected in any column.</span>
            </div>
        """))
    else:
        display(HTML(f"""
            <div style="
              padding: 12px 18px;
              background: #fdecea;
              border-left: 5px solid #e57373;
              border-radius: 8px;
              margin: 8px 0;
              font-size: 13.5px;
              color: #7b241c;
              font-weight: 600;
              line-height: 1.6;
              font-family: 'Inter', sans-serif;
            ">
                🚨 <strong>MISSING VALUES</strong><br>
                <span style="font-weight: 400;">Missing values detected in <strong>{len(report)}</strong> column(s).</span>
            </div>
        """))
        display(
            report.style
                  .background_gradient(cmap='PuBu', subset=['Percentage (%)'])
                  .format({'Percentage (%)': '{:.2f}%'})
                  .set_table_styles([{
                      'selector': 'table',
                      'props': [('width', '100%'), ('table-layout', 'auto')]
                  }])
        )

    n_dupes = df.duplicated().sum()
    if n_dupes == 0:
        display(HTML("""
            <div style="
              padding: 12px 18px;
              background: #e8f4fb;
              border-left: 5px solid #8CAEC2;
              border-radius: 8px;
              margin: 8px 0;
              font-size: 13.5px;
              color: #2c5f7a;
              font-weight: 600;
              line-height: 1.6;
              font-family: 'Inter', sans-serif;
            ">
                ✅ <strong>DUPLICATES</strong><br>
                <span style="font-weight: 400;">No duplicate rows detected.</span>
            </div>
        """))
    else:
        pct_dupes = round((n_dupes / len(df)) * 100, 2)
        display(HTML(f"""
            <div style="
              padding: 12px 18px;
              background: #fdecea;
              border-left: 5px solid #e57373;
              border-radius: 8px;
              margin: 8px 0;
              font-size: 13.5px;
              color: #7b241c;
              font-weight: 600;
              line-height: 1.6;
              font-family: 'Inter', sans-serif;
            ">
                🚨 <strong>DUPLICATES</strong><br>
                <span style="font-weight: 400;"><strong>{n_dupes}</strong> duplicate row(s) found — {pct_dupes}% of the dataset.</span>
            </div>
        """))

    return report
```

```python
def plot_categorical_distributions(df, cols=None, target=None):
  
    if cols is None:
        cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
    if not cols:
        display(HTML("<p>No categorical columns found to plot.</p>"))
        return
    display(HTML("""
        <div style="
          padding: 14px 26px;
          background: linear-gradient(90deg, #8CAEC2, #BB8CC2);
          border-radius: 12px;
          border-left: 7px solid lightblue;
          color: white;
          font-family: 'Inter', sans-serif;
          margin: 42px 0 18px;
          box-shadow: 0 4px 16px rgba(0,0,0,0.08);
          display: flex;
          align-items: center;
          gap: 14px;
        ">
          <h3 style="font-size: 18px; font-weight: 700; margin: 0; color: white;">📊 Categorical Distributions</h3>
        </div>
    """))
    for col in cols:
        if target is not None and col != target:
            fig, axes = plt.subplots(1, 2, figsize=(16, 5))

            order = df[col].value_counts().index

            ax1 = sns.countplot(data=df, y=col, order=order, palette='GnBu', hue=col, legend=False, ax=axes[0])
            axes[0].set_title(f"Frequency Distribution: {col}", fontsize=14, loc='left', pad=10)
            axes[0].set_xlabel("Count", fontsize=12)
            axes[0].set_ylabel(col, fontsize=12)
            for p in ax1.patches:
                width = p.get_width()
                ax1.text(width + (width * 0.01), p.get_y() + p.get_height()/2,
                         f'{int(width)}', ha='left', va='center')
            sns.despine(ax=axes[0])

            ax2 = sns.countplot(data=df, y=col, order=order, hue=target, palette=['#2c5f7a', '#BB8CC2'], ax=axes[1])
            axes[1].set_title(f"{col} by {target}", fontsize=14, loc='left', pad=10)
            axes[1].set_xlabel("Count", fontsize=12)
            axes[1].set_ylabel(col, fontsize=12)
            axes[1].legend(title=target, bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=10)
            for p in ax2.patches:
                width = p.get_width()
                ax2.text(width + (width * 0.01), p.get_y() + p.get_height()/2,
                         f'{int(width)}', ha='left', va='center')
            sns.despine(ax=axes[1])

            plt.tight_layout()
            plt.show()

        else:
            plt.figure(figsize=(10, 5))
            
            order = df[col].value_counts().index
            
            ax = sns.countplot(data=df, y=col, order=order, palette='GnBu', hue=col, legend=False)
            
            plt.title(f"Frequency Distribution: {col}", fontsize=14, loc='left', pad=10)
            plt.xlabel("Count", fontsize=12)
            plt.ylabel(col, fontsize=12)
            for p in ax.patches:
                width = p.get_width()
                ax.text(width + (width * 0.01), p.get_y() + p.get_height()/2, 
                        f'{int(width)}', ha='left', va='center')
            sns.despine()
            plt.tight_layout()
            plt.show()
```

        <li><strong>Ordinal encode Contract</strong> — preserves the natural commitment order: <em>Month-to-month → 0</em>, <em>One year → 1</em>, <em>Two year → 2</em>.</li>
        <li><strong>One-hot encode multi-class categoricals</strong> — applies <code style="background:#f0eaf8; color:#7b5ea7; padding:1px 5px; border-radius:4px;">pd.get_dummies</code> to <em>InternetService</em> and <em>PaymentMethod</em>, the only remaining features with more than two categories.</li>
        <li><strong>Scale numeric features</strong> — applies <code style="background:#f0eaf8; color:#7b5ea7; padding:1px 5px; border-radius:4px;">StandardScaler</code> to <em>tenure</em>, <em>MonthlyCharges</em>, and <em>TotalCharges</em>. The fitted scaler is saved and reused on the test set to prevent data leakage.</li>
    </ol>
</div>

```python
def preprocess(df, is_train=True, scaler=None, train_columns=None):

    df = df.copy()
    mode_label = "🟦 TRAIN" if is_train else "🟪 TEST"

    display(HTML(f"""
        <div style="padding: 10px 18px; background: #f0f7ff; border-left: 5px solid #BB8CC2;
                    border-radius: 8px; margin: 6px 0 14px; font-size: 13px; color: #5a3e7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            {mode_label} &nbsp;— Processing {df.shape[0]:,} rows × {df.shape[1]} columns
        </div>
    """))

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            1️⃣ &nbsp;Encode Target Variable (Churn)
        </div>
    """))
    if 'Churn' in df.columns:
        df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
        print(f"   → Churn encoded: Yes → 1 | No → 0")
        print(f"   → Class balance:\n{df['Churn'].value_counts(normalize=True).rename({1:'Churn=1', 0:'Churn=0'}).to_string()}")
    else:
        print("   → Churn column not present (test set) — skipped")

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            2️⃣ &nbsp;Simplify 'No internet / phone service' → 'No'
        </div>
    """))
    service_cols = [
        'MultipleLines', 'OnlineSecurity', 'OnlineBackup',
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
    ]
    for col in service_cols:
        df[col] = df[col].replace({'No internet service': 'No', 'No phone service': 'No'})
    print(f"   → Replaced in: {service_cols}")

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            3️⃣ &nbsp;Binary Encode Yes / No Columns
        </div>
    """))
    binary_cols = [
        'gender', 'Partner', 'Dependents', 'PhoneService',
        'PaperlessBilling'
    ] + service_cols
    binary_map = {'Yes': 1, 'No': 0, 'Male': 1, 'Female': 0}
    for col in binary_cols:
        df[col] = df[col].map(binary_map)
    print(f"   → Binary encoded: {binary_cols}")

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            4️⃣ &nbsp;Ordinal Encode Contract
        </div>
    """))
    contract_map = {'Month-to-month': 0, 'One year': 1, 'Two year': 2}
    df['Contract'] = df['Contract'].map(contract_map)
    print(f"   → Contract encoded: Month-to-month → 0 | One year → 1 | Two year → 2")

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            5️⃣ &nbsp;One-Hot Encode MultiClass Categoricals
        </div>
    """))
    ohe_cols = ['InternetService', 'PaymentMethod']
    df = pd.get_dummies(df, columns=ohe_cols, drop_first=False, dtype=int)
    new_cols = [c for c in df.columns if any(c.startswith(p) for p in ohe_cols)]
    print(f"   → One-hot encoded: {ohe_cols}")
    print(f"   → New columns created: {new_cols}")

    if not is_train and train_columns is not None:
        target_col = 'Churn' if 'Churn' in df.columns else None
        feature_cols = [c for c in train_columns if c != 'Churn']
        for col in feature_cols:
            if col not in df.columns:
                df[col] = 0
        df = df[[c for c in train_columns if c in df.columns]]
        print(f"   → Columns aligned to training set shape: {df.shape[1]}")

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            6️⃣ &nbsp;Scale Numeric Features
        </div>
    """))
    scale_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
    if is_train:
        scaler = StandardScaler()
        df[scale_cols] = scaler.fit_transform(df[scale_cols])
        print(f"   → StandardScaler fitted and applied to: {scale_cols}")
    else:
        df[scale_cols] = scaler.transform(df[scale_cols])
        print(f"   → Fitted StandardScaler applied to: {scale_cols}")

    display(HTML(f"""
        <div style="
          padding: 16px 22px;
          background: #e8f4fb;
          border-left: 5px solid #BB8CC2;
          border-radius: 10px;
          margin: 18px 0 6px;
          font-family: 'Inter', sans-serif;
          font-size: 13.5px;
          color: #3a3a3a;
          line-height: 1.8;
          box-shadow: 0 2px 10px rgba(140,174,194,0.12);
        ">
          <strong style="color: #2c5f7a; font-size: 14px;">✅ Preprocessing Complete — {mode_label}</strong><br>
          Shape after preprocessing: <strong>{df.shape[0]:,} rows × {df.shape[1]} columns</strong><br>
          Remaining object columns: <strong>{list(df.select_dtypes(include='object').columns) or 'None'}</strong><br>
          Remaining nulls: <strong>{df.isnull().sum().sum()}</strong>
        </div>
    """))

    if is_train:
        return df, scaler, df.columns.tolist()
    else:
        return (df,)
```

```python
train_df_processed, scaler, train_columns = preprocess(train_df, is_train=True)
```

```python
 def plot_target_correlation(df, target='Churn', figsize=(12, 7)):

    display(HTML("""
        <div style="
          padding: 14px 26px;
          background: linear-gradient(90deg, #8CAEC2, #BB8CC2);
          border-radius: 12px;
          border-left: 7px solid lightblue;
          color: white;
          font-family: 'Inter', sans-serif;
          margin: 42px 0 18px;
          box-shadow: 0 4px 16px rgba(0,0,0,0.08);
          display: flex;
          align-items: center;
          gap: 14px;
        ">
          <h3 style="font-size: 18px; font-weight: 700; margin: 0; color: white;">🎯 Target Correlation Analysis</h3>
        </div>
    """))

    corr = (
        df.corr(numeric_only=True)[target]
        .drop(labels=[target])
        .sort_values(key=abs, ascending=False)
    )

    colors = ['#2c5f7a' if v >= 0 else '#BB8CC2' for v in corr.values]

    fig, ax = plt.subplots(figsize=figsize)

    bars = ax.barh(corr.index[::-1], corr.values[::-1], color=colors[::-1],
                   edgecolor='white', linewidth=0.6)

    for bar, val in zip(bars, corr.values[::-1]):
        ax.text(
            val + (0.003 if val >= 0 else -0.003),
            bar.get_y() + bar.get_height() / 2,
            f'{val:.3f}',
            va='center',
            ha='left' if val >= 0 else 'right',
            fontsize=9,
            color='#3a3a3a'
        )

    ax.axvline(0, color='#8CAEC2', linewidth=1.2, linestyle='--', alpha=0.7)
    ax.set_xlabel("Pearson Correlation Coefficient", fontsize=11, color='#3a3a3a')
    ax.set_title(f"Feature Correlation with '{target}'", fontsize=14, loc='left',
                 pad=12, color='#2c5f7a', fontweight='bold')
    ax.tick_params(axis='y', labelsize=10)
    ax.tick_params(axis='x', labelsize=9)
    sns.despine()
    plt.tight_layout()
    plt.show()

    display(HTML("""
        <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                    border-radius: 8px; margin: 18px 0 8px; font-size: 13px; color: #2c5f7a;
                    font-weight: 700; font-family: 'Inter', sans-serif;">
            📋 Correlation Table
        </div>
    """))

    corr_df = corr.reset_index()
    corr_df.columns = ['Feature', f'Correlation with {target}']
    corr_df['Abs Correlation'] = corr_df[f'Correlation with {target}'].abs().round(4)
    corr_df[f'Correlation with {target}'] = corr_df[f'Correlation with {target}'].round(4)
    corr_df = corr_df.set_index('Feature')

    display(
        corr_df.style
               .background_gradient(cmap='PuBu', subset=['Abs Correlation'])
               .bar(subset=[f'Correlation with {target}'], color=['#BB8CC2', '#2c5f7a'],
                    align='zero', width=80)
               .format({f'Correlation with {target}': '{:+.4f}', 'Abs Correlation': '{:.4f}'})
               .set_table_styles([{
                   'selector': 'table',
                   'props': [('width', '100%'), ('table-layout', 'auto')]
               }])
    )
```

```python
TARGET = 'Churn'
X = train_df_processed.drop(columns=[TARGET])
y = train_df_processed[TARGET]

display(HTML("""
    <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                font-weight: 700; font-family: 'Inter', sans-serif;">
        1️⃣ &nbsp;Model Definition
    </div>
"""))

model = XGBClassifier(
    n_estimators=2000,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='auc',
    random_state=42,
    n_jobs=-1
)

print(f"   → XGBClassifier initialised")
print(f"   → n_estimators: 2000 | learning_rate: 0.05 | max_depth: 6")
print(f"   → subsample: 0.8 | colsample_bytree: 0.8")
```

```python
display(HTML("""
    <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                font-weight: 700; font-family: 'Inter', sans-serif;">
        2️⃣ &nbsp;Stratified K-Fold Cross-Validation (ROC-AUC)
    </div>
"""))

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc', n_jobs=-1)

print(f"   → Fold scores: {[round(s, 4) for s in cv_scores]}")
print(f"   → Mean ROC-AUC : {cv_scores.mean():.4f}")
print(f"   → Std  ROC-AUC : {cv_scores.std():.4f}")

fig, ax = plt.subplots(figsize=(8, 3.5))
bars = ax.bar(
    [f"Fold {i+1}" for i in range(len(cv_scores))],
    cv_scores,
    color=['#2c5f7a' if s >= cv_scores.mean() else '#BB8CC2' for s in cv_scores],
    edgecolor='white', linewidth=0.8, width=0.55
)
ax.axhline(cv_scores.mean(), color='#8CAEC2', linestyle='--', linewidth=1.5,
           label=f'Mean AUC = {cv_scores.mean():.4f}')
for bar, val in zip(bars, cv_scores):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.002,
            f'{val:.4f}', ha='center', va='bottom', fontsize=10, color='#3a3a3a')
ax.set_ylim(min(cv_scores) - 0.02, 1.0)
ax.set_ylabel("ROC-AUC", fontsize=11)
ax.set_title("Cross-Validation ROC-AUC per Fold", fontsize=13, loc='left',
             pad=10, color='#2c5f7a', fontweight='bold')
ax.legend(fontsize=10)
sns.despine()
plt.tight_layout()
plt.show()

display(HTML("""
    <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                font-weight: 700; font-family: 'Inter', sans-serif;">
        3️⃣ &nbsp;Out-of-Fold Predictions & ROC Curve
    </div>
"""))

oof_proba = cross_val_predict(model, X, y, cv=cv, method='predict_proba', n_jobs=-1)[:, 1]
oof_pred  = (oof_proba >= 0.5).astype(int)
oof_auc   = roc_auc_score(y, oof_proba)

fpr, tpr, _ = roc_curve(y, oof_proba)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(fpr, tpr, color='#2c5f7a', lw=2, label=f'OOF ROC Curve (AUC = {oof_auc:.4f})')
ax.plot([0, 1], [0, 1], color='#BB8CC2', lw=1.2, linestyle='--', label='Random Classifier')
ax.fill_between(fpr, tpr, alpha=0.07, color='#8CAEC2')
ax.set_xlabel("False Positive Rate", fontsize=11)
ax.set_ylabel("True Positive Rate", fontsize=11)
ax.set_title("Out-of-Fold ROC Curve", fontsize=13, loc='left',
             pad=10, color='#2c5f7a', fontweight='bold')
ax.legend(fontsize=10)
sns.despine()
plt.tight_layout()
plt.show()

print(f"   → OOF ROC-AUC: {oof_auc:.4f}")

display(HTML("""
    <div style="padding: 12px 18px; background: #e8f4fb; border-left: 5px solid #8CAEC2;
                border-radius: 8px; margin: 10px 0 6px; font-size: 13px; color: #2c5f7a;
                font-weight: 700; font-family: 'Inter', sans-serif;">
        4️⃣ &nbsp;Classification Report & Confusion Matrix
    </div>
"""))

print(classification_report(y, oof_pred, target_names=['No Churn', 'Churn']))

fig, ax = plt.subplots(figsize=(5, 4))
cm = confusion_matrix(y, oof_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Churn', 'Churn'])
disp.plot(ax=ax, colorbar=False, cmap='PuBu')
ax.set_title("Confusion Matrix (OOF)", fontsize=13, loc='left',
             pad=10, color='#2c5f7a', fontweight='bold')
plt.tight_layout()
plt.show()
```
